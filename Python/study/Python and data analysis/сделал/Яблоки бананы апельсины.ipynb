1.Исключим одну переменную
---

Преобразуем уравнение:

$\frac{a}{b+c} + \frac{b}{a+c} + \frac{c}{a+b} = 4$

в:

$\frac{x}{y+1} + \frac{y}{x+1} + \frac{1}{x+y} = 4$

Перемножим на знаменатели и получим:

$x(x+1)(x+y) + y(y+1)(x+y) + (x+1)(y+1) = 4(x+1)(y+1)(x+y)$

Расскрыв скобки и перенеся все в правую часть мы получим уравнение:

$x^3 - 3yx^2 - 3x^2 - 3xy^2 - 5xy - 3x + y^3 - 3y^2 - 3y + 1 = 0$

3.Определиv третью точку пересечения
---

Если у нас есть прямая $y=ax+b$, то мы можем исключить переменную $y$ из уравнения кривой. Заменив $y$ на $(ax+b)$ и переупорядочив его, мы получим кубический полиномиал:

$(a^3-3a^2-3a+1)x^3 + (3a^2b-3a^2-6ab-5a-3b-3)x^2 + (3ab^2-6ab-3a-3b^2-5b-3)x+b^3-3b^2-3b+1=0$

Соеденим точки $(x_{1}, y_{1})$ и $(x_{2}, y_{2})$ и получим уравнение прямой

$$y=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}(x-x_{1})+y_{1}$$

Для точек $P(-4, -11)$ и $Q(\frac{11}{4}, \frac{-1}{4}), $ получим линейное уравнение $y=\frac{43}{27}x-\frac{125}{27}$, то есть $a=\frac{43}{27}$ и $b=\frac{125}{27}$ для кубического уравнения, получим:

$$-20660x^3+12171x^2+274755x-417956=0$$

Теперь у нас есть кубическое уравнение, которое мы можем решить. Выполним длинное деление для каждого известного корня: мы делим наш полином на $(x-x_i)$, где $x_i$ - это корень. Итак, первое деление на $(x+4)$:

$$-20660x^2+94811x-104489=0$$

Потом разделим на $(x-\frac{11}{4})$ и упростим:

$$-5165x+9499=0$$

Из этого линейного уравнения получаем $x=\frac{9499}{5165}$. $y$ координату получим, вычисляя уравнение линии на $x$, то есть $y=-\frac{8784}{5165}$. Получим точку пересечения $R(\frac{9499}{5165}, -\frac{8784}{5165})$. Теперь поменяем координаты $x$ и $y$ и получим нашу следующую рациональную точку $R'(-\frac{8784}{5165}, \frac{9499}{5165})$ на кривой.

5.Код на питоне
---

```python
from fractions import Fraction

def line(x0, y0, x1, y1):
    """
    Сделаем полином, который является уравнением прямой, проходящей через точки (x0,y0) и (x1,y1).
    """
    return [(y1-y0)/(x1-x0), y0-(y1-y0)/(x1-x0)*x0]

def intersect_curve(a, b):
    """
    Возвращает полином, который является пересечением эллиптической кривой и линии ax+b:
    (a^3-3a^2-3a+1)*x^3 + (3a^2*b-3a^2-6ab-5a-3b-3)*x^2 + (3ab^2-6ab-3a-3b^2-5b-3)*x + b^3-3b^2-3b+1 = 0
    """
    return [(a*a*a-3*a*a-3*a+1), (3*a*a*b-3*a*a-6*a*b-5*a-3*b-3), (3*a*b*b-6*a*b-3*a-3*b*b-5*b-3), b*b*b-3*b*b-3*b+1]

def evaluate(p, x):
    """
    Оценим полиномиальное p для значения x
    """
    y = 0
    first = True
    for c in p:
        if first:
            first = False
        else:
            y *= x
        y += c
    return y

def divide(p1, p2):
    """
    Разделим два полинома и вернем коэффициент.
    """
    result = []
    remainder = p1.copy()
    for i in range(len(p1)-len(p2)+1):
        r = remainder[i]/p2[0]
        result.append(r)
        for j in range(len(p2)):
            if i+j >= len(p1):
                break
            remainder[i+j] = remainder[i+j] - r*p2[j]
    return result

# (x1, y1) и (x2, y2) - это две рациональные точки на эллиптической кривой
x1 = Fraction(-4,1)
y1 = Fraction(-11,1)
x2 = Fraction(11,4)
y2 = Fraction(-1,4)

while True:
    # сделаем хорду, которая пройдет через наши две рациональные точки.
    chord = line(x1, y1, x2, y2)

    # пересекает эллиптическую кривую по аккорду.
    # точки пересечения - это корни полинома пересечения
    cubic = intersect_curve(chord[0], chord[1])

    # разделим кубический полином пересечения на (x-x1), где x1 - один корень
    # коэффициент будет квадратичным полиномом.
    quadratic = divide(cubic, [Fraction(1,1), -x1])

    # снова делим квадратичный полином на (x-x2), где x2 - это другой корень.
    # коэффициент будет линейным многочленом, чей корень - третий корень.
    linear = divide(quadratic, [Fraction(1,1), -x2])

    # третий корень определяет третью точку пересечения кривой и хорды
    x3 = -linear[1] / linear[0]
    y3 = evaluate(chord, x3)
    if x3 > 0 and y3 > 0:
        break

    # отобразить третью точку пересечения через линию y=x путём замены x и y
    # это будет наш следующий рациональной точкой, который мы будем использовать для следующей итерации.
    x2 = y3
    y2 = x3

a = x3.numerator
b = y3.numerator
c = x3.denominator

print('Решение:')
print('A =', a)
print('B =', b)
print('C =', c)
print('A/(B+C) + B/(A+C) + C/(A+B) =', Fraction(a, b+c) + Fraction(b, a+c) + Fraction(c, a+b))

```
