scipy.optimize.golden() - находит минимум функции. Как было показано в конце предыдущего шага, если не передавать в неё никаких параметров кроме функции, минимум которой ищется, то возвращён будет аргумент, при котором этот минимум достигается. Примечание. Пример дан для $a=3$, $b=4$ 

```python
from math import fabs
from scipy.optimize import golden


a = 3
b = 4

def f(x):
    return (x + a)**2 - b


def g(x):
    return fabs(f(x))


min_f = golden(f)
min_g = golden(g, full_output=True)

print(min_f, min_g)
```

Научимся работать с параметрами функций оптимизации на примере scipy.optimize.golden().

Рассмотрим 2 параметра:

- brack - первичное ограничение аргумента для поиска минимума (найденное решение может быть за пределами интервала, однако имеет смысл использовать его, например, для предотвращения скатывания в локальный минимум)
- full_output - если равен True, то функция возвращает кортеж вида (аргумент_минимума, значение_минимума_функции, число итераций)

Пример:

    return (x + 10)**2

```python
from math import fabs
from scipy.optimize import golden


a = 3
b = 4

def f(x):
    return (x + a)**2 - b


def g(x):
    return fabs(f(x))


arg_f, val_f, iter_f = golden(f, brack=(-10, -4), full_output=True)
arg_g, val_g, iter_g = golden(g, brack=(-10, -4), full_output=True)

print(val_f, val_g)
```

При использовании параметра full_output=True функция golden() вернет следующий кортеж:

(<значения аргумента функции в точке минимума>,

 <значение самой функции в точке минимума>,
 
 <число итераций понадобившихся для нахождения минимума>)
