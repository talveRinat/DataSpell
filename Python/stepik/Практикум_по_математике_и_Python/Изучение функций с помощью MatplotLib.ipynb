Цель задания - исследование функции

$\LARGE f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$

при различных значениях $\alpha$ и $\beta$.

Для каждой задачи необходимо:

- Построить график (размер графика должен быть достаточным, чтобы визуально увидеть особенности изучаемых функций), график каждой функции должен быть одного цвета для одного значения α \alpha α и β \beta β.
- Подписать оси и заголовок
- Создать легенду
- Сохранить изображение в svg файл
- Сделать и записать выводы о поведении графиков (можно использовать наводящие вопросы)



- Код не должен вызывать ошибки исполнения (например, из-за деления на 0 или корня из отрицательной величины)
- Код должен выполняться (падение кода с ошибкой у проверяющего на python аналогичной версии - повод не зачесть все пункты соответствующей задачи)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
xpl=np.arange(0.01,500,0.01) #положительные х

xotr=np.arange(-0.01,-500,-0.01) #отрицательные х
```

```python
def f(x,alfa,beta):
    f=(x**beta + alfa**beta)/x**beta
    return f
```

# Задача 1. Часть 1.

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$ 
- $\alpha = 2, \beta = 1$ 
- $\alpha = 1, \beta = 2$


```python
alfa1 = 1
beta1 = 1

alfa2 = 2
beta2 = 1

alfa3 = 1
beta3 = 2
```

```python
f1pl = f(xpl,alfa1,beta1)
f1otr = f(xotr,alfa1,beta1)

f2pl = f(xpl,alfa2,beta2)
f2otr = f(xotr,alfa2,beta2)

f3pl = f(xpl,alfa3,beta3)
f3otr = f(xotr,alfa3,beta3)
```

```python
fig1 = plt.figure(figsize=(13,8),dpi=120)

plt.ylim(-20,20)
plt.xlim(-5,5)

plt.plot(xpl, f1pl, 'b', label='функция f(x) при $\\alpha=1, \\beta=1$')
plt.plot(xotr, f1otr, 'b')

plt.plot(xpl, f2pl, 'r', label='функция f(x) при $\\alpha=2, \\beta=1$')
plt.plot(xotr, f2otr, 'r')

plt.plot(xpl, f3pl, 'g', label='функция f(x) при $\\alpha=1, \\beta=2$')
plt.plot(xotr, f3otr, 'g')

plt.xlabel('x')

plt.ylabel('f(x)')

plt.title('График функции $ f(x)=\\frac{x^\\alpha + \\alpha^\\beta}{x^\\beta} $ \n \
при различных значениях параметров $\\alpha \:и\: \\beta$')

plt.legend()
plt.show()
plt.savefig('fig1.svg')
```
