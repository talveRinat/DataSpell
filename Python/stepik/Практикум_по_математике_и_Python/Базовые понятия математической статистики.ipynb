На вход подаётся строка из натуральных чисел, разделённых пробелами.

Сформируйте отчёт о мерах центральной тенденции данной выборки (Среднем, Медиане, Моде/Модах) и соотношениях между ними.

Отчёт должен иметь вид:

    1 строка - "Среднее: X" (где x - значение среднего арифметического для выборки, округлённое до 1 знака после запятой)
    2 строка - "Медиана: Y" (где y - значение медианы для выборки, округлённое до 1 знака после запятой)

 

Если значение моды единственное:

    3 строка - "Мода: Z"

Если значений мод несколько:

    3 строка - "Моды: Z1 Z2 Z3 ... Zn" (где Z1 Z2 Z3 ... Zn - значения мод, по возрастанию, записанные через пробел, рекомендуется использовать функцию из первой задачи)

 

    4 и последующие строки - строки, описывающие соотношения Мод, Медианы и Среднего (в произвольном порядке, главное чтобы 1 соотношение было записано на 1 строке).


```python
import numpy as np 
import pandas as pd


n = list(map(int, input().split()))
Mo =  []
for i in pd.Series(n).mode():
    Mo.append(i)
Me = np.median(n)
Cp = np.mean(n)

d = {
    1 : 'Ме<Ср',
    2 : 'Ме>Ср',
    3 : 'Ме=Ср',
    4 : 'Мо<Ср',
    5 : 'Мо>Ср',
    6 : 'Мо=Ср',
    7 : 'Мо>Ме',
    8 : 'Мо<Ме',
    9 : 'Мо=Ме'
}

lst = []
if np.any(Me<Cp):
    lst.append(1)
if np.any(Me>Cp):
    lst.append(2)
if np.any(Me==Cp):
    lst.append(3)
if np.any(Mo<Cp):
    lst.append(4)
if np.any(Mo>Cp):
    lst.append(5)
if np.any(Mo==Cp):
    lst.append(6)
if np.any(Mo>Me):
    lst.append(7)
if np.any(Mo<Me):
    lst.append(8)
if np.any(Mo==Me):
    lst.append(9)


if len(Mo) > 1:
    letter = 'ы'
else:
    letter = 'а'
    
print(f'Среднее: {round(Cp, 1)}')
print(f'Медиана: {round(Me, 1)}')
print(f'Мод{letter}: {" ".join(str(x) for x in Mo)}')
for i in (lst):
    print(d.get(i))
```

```python
import pandas as pd


data = pd.Series(map(int, input().split()))

aver = round(float(data.mean()), 1)
med = round(float(data.median()), 1)
mods = data.mode()

print(f'Среднее: {aver}')
print(f'Медиана: {med}')
print(f'Мод{"а" if len(mods) == 1 else "ы"}: ', end='')
print(*mods)

conditions = [med<aver, med>aver, med==aver, 
              min(mods)<aver, max(mods)>aver, any(aver == mod for mod in mods), 
              max(mods)>med, min(mods)<med, any(med == mod for mod in mods)]

print_out = ['Ме<Ср', 'Ме>Ср', 'Ме=Ср', 'Мо<Ср', 'Мо>Ср', 'Мо=Ср', 'Мо>Ме', 'Мо<Ме', 'Мо=Ме']

print(*[print_out[i] for i in range(9) if conditions[i]], sep='\n')
```

```python
import numpy
import pandas


lst=list(map(int,input().split()))

mean=numpy.mean(lst)
print('Среднее:',round(mean,1))

median=numpy.median(lst)
print('Медиана:',round(median,1))

moda=pandas.Series.mode(lst)
print('Мода:',*moda) if len(moda)==1 else print('Моды:',*moda)

def output(a,b,a1,b1):    
    if a>b:
        print('{}>{}'.format(a1,b1))
    elif a==b:
        print('{}={}'.format(a1,b1))
    else:
        print('{}<{}'.format(a1,b1))

output(median,mean,'Ме','Ср')
for elem in moda:
    output(elem,mean,'Мо','Ср')
    output(elem,median,'Мо','Ме')
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(1, 1, figsize=(15,8))

f = stats.exponnorm(K=5.5, loc=0, scale=0.5)

ax.grid()
x = np.linspace(f.ppf(0.0001), f.ppf(0.99), 100)

ax.plot([x[np.where(f.pdf(x)==max(f.pdf(x)))[0][0]], x[np.where(f.pdf(x)==max(f.pdf(x)))[0][0]]], 
        [0, max(f.pdf(x))], '-', lw=3)

plt.annotate('2', (x[np.where(f.pdf(x)==max(f.pdf(x)))[0][0]],0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.mean(), f.mean()], [0, f.pdf(f.mean())], '-y', lw=3)

plt.annotate('4', (f.mean(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.mean()+f.std(), f.mean()+f.std()], [0, f.pdf(f.mean()+f.std())], '-g', lw=3)

plt.annotate('5', (f.mean()+f.std(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.mean()-f.std(), f.mean()-f.std()], [0, f.pdf(f.mean()-f.std())], '-g', lw=3)

plt.annotate('1', (f.mean()-f.std(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.mean()+2*f.std(), f.mean()+2*f.std()], [0, f.pdf(f.mean()+2*f.std())], '-', lw=3)

plt.annotate('6', (f.mean()+2*f.std(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.mean()+3*f.std(), f.mean()+3*f.std()], [0, f.pdf(f.mean()+3*f.std())], '-', lw=3)

plt.annotate('7', (f.mean()+3*f.std(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')
ax.plot([f.median(), f.median()], [0, f.pdf(f.median())], '-', lw=3)

plt.annotate('3', (f.median(),0), textcoords="offset points", xytext=(0,-25), size=20, ha='center', color='red')

plt.xlim(f.ppf(0.0001), f.ppf(0.99))
plt.ylim(0, max(f.pdf(x))*1.1)
plt.tick_params(labelbottom=False, labelleft=False)
ax.plot(x, f.pdf(x), 'r-', lw=5, alpha=0.6, label='rice pdf')

plt.show()
```

Функция gen() возвращает 1 случайное число из некоторого распределения.

Вычислите выборочные среднее и стандартное отклонение и выведите их на печать по шаблону:

    Mean: x

    Sd: y

Значения среднего и стандартного отклонения округлите до 2 знака после запятой.

Как и в задачах на метод Монте-Карло, чем больше значений сгенерируете, тем точнее будет результат, однако в любом случае присутствует элемент случайности, так что он может отличаться от эталонного (теоретического), характерного для данного распределения.

1 распределение - stats.exponnorm(K=5.5, loc=0, scale=0.5)

2 распределение - stats.norm()

3 распределение - бросок кубика с 6 гранями

4 распределение - сумма очков при броске кубиков с 6 и 4 гранями

5 распределение - stats.dweibull(c = 2, loc=1, scale=0.5)

```python
data = np.array([gen() for _ in range(100000)])
mean = data.mean()
std = np.std(data, ddof=1)
print('Mean: {}\nSd: {}'.format(round(mean, 2), round(std, 2)))
```
