Дерево не может решить комплекстно, он всегда делаит на шаги.

Дереву решений очень трудно работать с изобраениям, слишком много данных так как объект может изображаться в разных позициях по пикселям.

**Дерево решений для номинативных фичей!**

Если данные низкоуровненевые(закрашен пиксель или нет)

Дерево действуей по методу разделяй и властвуй.




$$ln(\frac{p}{1-p})=w_1*X_1+w_2*X_2 ,$$

где $p$ - вероятность положительного события (наш объект - крестик), $1 - p$ - вероятность отрицательного события (наш объект - нолик). Отношение $\frac{p}{1-p}$, то есть положительного исхода к отрицательному, называется шанс. И от этого берется логарифм. Из этой формулы можно выразить $p$ - вероятность положительного исхода (то, что мы и хотим):

$$p=\frac{1}{1+exp^{-(w_1*X_1+w_2*X_2)}}$$

Эта функция ограничена сверху и снизу: область значений лежит на интервале $(0, 1)$.

Обозначим значение под экспонентой буквой $t$:

$$p=\frac{1}{1+exp^{-t}}$$

При $t=0:p=\frac{1}{1+1}=\frac{1}2$.

Теперь обратимся к собственно значению $t$. Это выражение можно посмотреть по-другому с точки линейно алгебры как на линейную комбинацию двух векторов. У нас есть 2 вектора $W$ и $X$ с координатами $W(w_1, w_2)$ и $X(X_1, X_2)$. Тогда их линейная комбинация по определению равна:

$W*X = w_1*X_1 + w_2*X_2$. И это именно то, что стоит под экспонентой!

Когда векторы перпендикулярны друг другу их линейная комбинация равна нулю! (тут надо связать формулу линейной комбинации с формулой скалярного произведения векторов). Перпендикулярность векторов приводит к значению $p = 0.5$. И теперь если у нас угол между векторами будет тупой, то произведение $W*X < 0$ и  $p < 0.5$. И наоборот при остром угле $W*X > 0$ и $p > 0.5$.

Мы хотим найти такой вектор $W(w_1, w_2)$, что для одних фичей-векторов $X_i(x_{i_1}, x_{i_2})$ у нас значение $p$ было $> 0.5$ (острый угол между $X_i$ и $W$). Такие фичи мы называем "положительными" исходами. 

А для других фичей-векторов $X_j(x_{j_1}, x_{j_2})$ значение $p$ было $< 0.5$ (тупой угол между $X_j$ и $W$). 

А потом мы "бросим" алгоритму неизвестный вектор $X_p(x_{p_1}, x_{p_2})$, про который мы не знаем, к "положительным" или "отрицательным" исходам его нужно отнести. А алгоритм будет "крутить" его и в конце определится.
